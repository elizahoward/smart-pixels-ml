```python
import OptimizedDataGenerator4 as ODG
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from pathlib import Path
import matplotlib.pyplot as plt
from config import BASE_DIR, TRAIN_DIR, VALIDATION_DIR

# 1) Define the improved XY‐profile model:
def CreateXYProfileModel():
    x_in = tf.keras.Input(shape=(21,1), name="x_profile")
    y_in = tf.keras.Input(shape=(13,1), name="y_profile")

    # x‐branch
    x = tf.keras.layers.Conv1D(32, 3, padding="same", activation="relu", name="conv_x")(x_in)
    x = tf.keras.layers.BatchNormalization(name="bn_x")(x)
    x = tf.keras.layers.GlobalMaxPooling1D(name="pool_x")(x)

    # y‐branch
    y = tf.keras.layers.Conv1D(16, 3, padding="same", activation="relu", name="conv_y")(y_in)
    y = tf.keras.layers.BatchNormalization(name="bn_y")(y)
    y = tf.keras.layers.GlobalMaxPooling1D(name="pool_y")(y)

    # merge + dense head
    merged = tf.keras.layers.Concatenate(name="concat_xy")([x, y])
    for units in [128, 64]:
        merged = tf.keras.layers.Dense(
            units,
            activation="relu",
            kernel_regularizer=tf.keras.regularizers.l2(1e-4),
            name=f"dense_{units}"
        )(merged)
        merged = tf.keras.layers.BatchNormalization(name=f"bn_{units}")(merged)
        merged = tf.keras.layers.Dropout(0.3, name=f"dropout_{units}")(merged)

    output = tf.keras.layers.Dense(
        1, activation="sigmoid",
        kernel_regularizer=tf.keras.regularizers.l2(1e-4),
        name="output"
    )(merged)

    model = tf.keras.Model(inputs=[x_in, y_in], outputs=output)
    model.compile(
        optimizer="adam",
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return model

# 2) Instantiate model and print summary
model = CreateXYProfileModel()
model.summary()

# 3) Paths to TFRecord folders
base_dir       = BASE_DIR
train_dir      = TRAIN_DIR
validation_dir = VALIDATION_DIR

# 4) Generators for x_profile & y_profile
train_gen = ODG.OptimizedDataGenerator(
    load_records=True,
    tf_records_dir=str(train_dir),
    x_feature_description=["x_profile", "y_profile"]
)
val_gen = ODG.OptimizedDataGenerator(
    load_records=True,
    tf_records_dir=str(validation_dir),
    x_feature_description=["x_profile", "y_profile"]
)

# 5) Callbacks: EarlyStopping + optional LR reduction
es = EarlyStopping(
    monitor="val_loss",
    patience=20,
    restore_best_weights=True
)
rlr = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=10,
    min_lr=1e-6,
    verbose=1
)

# 6) Train for 120 epochs
history = model.fit(
    x=train_gen,
    validation_data=val_gen,
    epochs=120,
    callbacks=[es, rlr],
    verbose=1,
    shuffle=False
)

# 7) Plot accuracy & loss
plt.figure(figsize=(6,4))
plt.plot(history.history["accuracy"], label="train acc")
plt.plot(history.history["val_accuracy"], label="val acc")
plt.xlabel("Epoch"); plt.ylabel("Accuracy")
plt.legend(); plt.grid("--", linewidth=0.5)
plt.title("Training vs. Validation Accuracy")
plt.show()

plt.figure(figsize=(6,4))
plt.plot(history.history["loss"], label="train loss")
plt.plot(history.history["val_loss"], label="val loss")
plt.xlabel("Epoch"); plt.ylabel("Binary Crossentropy")
plt.legend(); plt.grid("--", linewidth=0.5)
plt.title("Training vs. Validation Loss")
plt.show()
```
